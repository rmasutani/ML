```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy

np.random.seed(101)
sns.set()
plt.rcParams['figure.figsize'] = (11.0, 8.0)
```

## 棄却法を用いた乱数生成
棄却法を用いてベータ分布$Be(\alpha, \beta), (\alpha, \beta > 1)$に従う乱数を発生させてみる

```python
# Number of iterations
n_iter = 10000

# Beta distribution parameters
alpha = 2.5
beta = 3.0

R = ((alpha-1)/(alpha+beta-2))**(alpha-1) * ((beta-1)/(alpha+beta-2))**(beta-1) * (1/scipy.special.beta(alpha, beta)) 
accepted = []

for _ in range(n_iter):
    y = np.random.rand()
    u = np.random.rand()
    p_y = scipy.stats.beta.pdf(y, alpha, beta)
    while u > (1/R) * p_y:
        y = np.random.rand()
        u = np.random.rand()
        
        p_y = scipy.stats.beta.pdf(y, alpha, beta)
    
    accepted.append(y)
```

```python
plt.hist(accepted, 100, density=True)

xs = np.linspace(0, 1, 1000)
ys = scipy.stats.beta.pdf(xs, 2.5, 3.0)
plt.plot(xs, ys, color="r")
```
