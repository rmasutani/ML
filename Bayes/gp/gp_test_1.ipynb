```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(101)
```

```python
def gaussian_kernel(x1: np.array, x2: np.array, theta_1: int = 1, theta_2: int = 1) -> np.float64:
    return theta_1 * np.exp(-1*np.linalg.norm(x1-x2)**2 / theta_2)
```

```python
def make_gram_matrix(x: np.array) -> np.array:
    size = len(x)
    K = np.zeros(shape=(size, size))
    
    for i in range(size):
        for j in range(size):
            K[i, j] = gaussian_kernel(x[i], x[j])
    
    return K
```

```python
xs = np.linspace(-4, 4, 100)
size = len(xs)

loc = np.zeros(size)
scale = make_gram_matrix(xs)
```

```python
# Some samples from gaussian process
color = ["g","r","b","y", "m"]
for i in range(5):
    ys = np.random.multivariate_normal(loc, scale)
    plt.plot(xs, ys, color[i % len(color)])
```
